```python
x = torch.tensor([2.0], requires_grad=True)

# First computation
y1 = x ** 2
y1.backward()
print("After first backward:", x.grad.item())

# Trying to call .backward() again will fail:
try:
    y1.backward()
except Exception as e:
    print(f'(Exception: {e})')

# Retaining the graph allows us to backwards many times.
x = torch.tensor([2.0], requires_grad=True)
y1 = x ** 2
y1.backward(retain_graph=True)
print("After first backward:", x.grad.item())
y1.backward()
print("After second backward:", x.grad.item())

# Gradients are accumulating. Even if you define a new function of x.
y2 = x ** 2
y2.backward(retain_graph=True)
print("After third backward (over new function of x):", x.grad.item())

# Zero the gradients
x.grad.zero_()
print("After zeroing:", x.grad.item())

# Now compute again
y2.backward()
print("After third backward (after zeroing):", x.grad.item())
```

```python
import torch.optim as optim
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms

# Define data transformations
transform = transforms.Compose([
    transforms.ToTensor(),  # Convert PIL image to tensor
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalize to [-1, 1]
])

# Load CIFAR-10 dataset
trainset = torchvision.datasets.CIFAR10(
    root='./data', 
    train=True, 
    download=True, 
    transform=transform
)

# Get some random training images
trainloader = DataLoader(
    trainset, 
    batch_size=10, 
    shuffle=True, 
    num_workers=2
)

dataiter = iter(trainloader)
images, labels = next(dataiter)

print("Image batch shape:", images.shape)
print("Labels:", labels)
print("Label names:", [trainset.classes[label] for label in labels])
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # Convolutional layers
        self.conv1 = nn.Conv2d(3, 6, 5)  # 3 input channels, 6 output channels, 5x5 kernel
        self.pool = nn.MaxPool2d(2, 2)   # 2x2 max pooling
        self.conv2 = nn.Conv2d(6, 16, 5)  # 6 input channels, 16 output channels, 5x5 kernel
        
        # Fully connected layers
        self.fc1 = nn.Linear(16 * 5 * 5, 120)  # After conv and pooling, image is 5x5
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)  # 10 classes for CIFAR-10
    
    def forward(self, x):
        # First conv block
        x = self.pool(F.relu(self.conv1(x)))  # Conv -> ReLU -> Pool
        # Second conv block
        x = self.pool(F.relu(self.conv2(x)))  # Conv -> ReLU -> Pool
        
        # Flatten for fully connected layers
        x = x.view(-1, 16 * 5 * 5)
        
        # Fully connected layers
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)  # No activation on output (we'll use CrossEntropyLoss)
        return x

# Create the network
net = CNN()
print(net)
```

```python
# Loss function: CrossEntropyLoss is commonly used for classification
criterion = nn.CrossEntropyLoss()

# Optimizer: Stochastic Gradient Descent (SGD)
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)

# Alternative optimizers you might use:
# optimizer = optim.Adam(net.parameters(), lr=0.001)
# optimizer = optim.RMSprop(net.parameters(), lr=0.001)

print("Loss function:", criterion)
print("Optimizer:", optimizer)
```

```python
# Training loop for 10 epochs
for epoch in range(10):  # Loop over the dataset multiple times
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        # Get inputs and labels
        inputs, labels = data

        # EXERCISE: Zero the parameter gradients
        # 解释: 每次训练前必须把上一次算的梯度清零，否则梯度会累加导致错误
        optimizer.zero_grad()  # TODO: Fill this in

        # EXERCISE: Forward pass (pass inputs through the network)
        # 解释: 把图片喂给网络，得到预测结果
        outputs = net(inputs)  # TODO: Fill this in

        # EXERCISE: Compute loss (use criterion to compute loss between outputs and labels)
        # 解释: 计算预测结果(outputs)和真实标签(labels)之间的差距
        loss = criterion(outputs, labels)  # TODO: Fill this in

        # EXERCISE: Backward pass (call backward on loss)
        # 解释: 反向传播，算出每个参数对误差的"贡献"（梯度）
        loss.backward()  # TODO: Fill this in

        # EXERCISE: Update weights
        # 解释: 根据算出的梯度，更新网络参数
        optimizer.step()  # TODO: Fill this in

        # Print statistics
        running_loss += loss.item()
        if i % 2000 == 1999:    # Print every 2000 mini-batches
            print(f'[Epoch {epoch + 1}, Batch {i + 1}] Loss: {running_loss / 2000:.3f}')
            running_loss = 0.0

print('Finished Training')
```

```python
from datasets import load_dataset
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Load a financial dataset from HuggingFace
try:
    # Load sebdg/trading_data dataset from HuggingFace
    # Dataset: https://huggingface.co/datasets/sebdg/trading_data
    dataset = load_dataset('sebdg/trading_data', split='train[:5000]')
    print("Dataset loaded successfully!")
    print(f"Dataset features: {dataset.features}")
except Exception as e:
    print(f"Error loading dataset: {e}")
    print("Please check your internet connection and SSL settings.")
    raise
```

```python
# Convert dataset to DataFrame if it's not already
if isinstance(dataset, pd.DataFrame):
    df = dataset.copy()
else:
    df = pd.DataFrame(dataset)

# The dataset has capitalized column names, so we'll normalize them to lowercase
# Columns: Open, High, Low, Close, Adj Close, Volume, File, Date
df.columns = df.columns.str.lower()


# Display basic information
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nColumn names:", df.columns.tolist())

# Calculate daily price features
# The dataset already contains daily OHLCV data
# We'll create additional features from the daily prices

# Daily price range (high - low)
df['daily_range'] = df['high'] - df['low']

# Daily price change (close - open)
df['daily_change'] = df['close'] - df['open']

# Daily return (percentage change from open to close)
df['daily_return'] = (df['close'] - df['open']) / df['open']

# Price position within the day's range (where close is relative to high-low)
df['price_position'] = (df['close'] - df['low']) / (df['high'] - df['low'] + 1e-8)  # Add small epsilon to avoid division by zero

# In real trading, bid = highest buy order, ask = lowest sell order

# Calculate price change and returns
# Sort by date to ensure chronological order (important for time series)
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
    df = df.sort_values('date').reset_index(drop=True)

df['price_change'] = df['close'].diff()

# Note: daily_return is already calculated above

# Calculate moving averages for trend features
df['ma_5'] = df['close'].rolling(window=5).mean()
df['ma_10'] = df['close'].rolling(window=10).mean()

# Create labels: Buy (1) if price will increase in next period, else 0
# We'll predict if the price will go up in the next time step
df['future_return'] = df['daily_return'].shift(-1)  # Next day's return
df['label'] = (df['future_return'] > 0).astype(int)  # 1 if price goes up, 0 otherwise

# Drop rows with NaN values (from rolling windows and shifts)
df = df.dropna().reset_index(drop=True)

print(f"\nProcessed dataset shape: {df.shape}")
print(f"Buy signals: {df['label'].sum()}, No-buy signals: {(df['label'] == 0).sum()}")
```

```python
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# EXERCISE: Define the Binary Cross Entropy loss function and Adam optimizer
# Hint: Use nn.BCELoss()
# 解释: 因为这是二分类问题(输出是0到1的概率)，且模型最后一层有Sigmoid，
# 所以必须用 BCELoss (Binary Cross Entropy Loss)。
criterion = nn.BCELoss()  # TODO: Fill this in

# Hint: Use optim.Adam with learning rate 0.001
# 解释: Adam 是目前最流行的优化算法，能自适应调整学习率。
# model.parameters() 告诉优化器："请帮我更新这个模型里所有的权重"。
optimizer = optim.Adam(model.parameters(), lr=0.001)  # TODO: Fill this in

# Training parameters
num_epochs = 50
batch_size = 32

# Create DataLoader for batching
# 解释: 把之前的 Tensor 封装成数据集，方便按批次读取
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Training loop
model.train()
for epoch in range(num_epochs):
    running_loss = 0.0
    
    for batch_X, batch_y in train_loader:
        # EXERCISE: Zero gradients
        # Hint: Use optimizer.zero_grad()
        # 解释: 清空上一轮的梯度缓存
        optimizer.zero_grad()  # TODO: Fill this in
        
        # EXERCISE: Forward pass
        # Hint: Pass batch_X through the model
        # 解释: 把这一批数据喂给模型
        outputs = model(batch_X)  # TODO: Fill this in
        
        # EXERCISE: Compute loss
        # Hint: Use criterion over outputs and batch_y
        # 解释: 算算预测值和真实值(batch_y)差多少
        loss = criterion(outputs, batch_y)  # TODO: Fill this in
        
        # EXERCISE: Backward pass and optimization
        # Hint: Use loss.backward() to compute gradients, then optimizer.step() to update weights
        # 解释: 反向传播算梯度 + 更新参数
        loss.backward()  # TODO: Fill this in
        optimizer.step() # TODO: Fill this in
        
        running_loss += loss.item()
        
    # Print loss every 10 epochs
    if (epoch + 1) % 10 == 0:
        avg_loss = running_loss / len(train_loader)
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {avg_loss:.4f}')

print('Training completed!')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. 准备数据 (无需 Tensor，直接用 DataFrame)
# ==========================================
# 确保取最后 20% 的数据作为回测
test_size = int(len(df) * 0.2)
sequential_data = df.iloc[-test_size:].copy().reset_index(drop=True)

# 确保 MA_5 和 MA_10 存在
if 'ma_5' not in sequential_data.columns or 'ma_10' not in sequential_data.columns:
    print("正在重新计算 MA 指标...")
    sequential_data['ma_5'] = sequential_data['close'].rolling(window=5).mean()
    sequential_data['ma_10'] = sequential_data['close'].rolling(window=10).mean()

# ==========================================
# 2. 初始化账户
# ==========================================
initial_capital = 10000.0
cash = initial_capital
shares = 0
portfolio_values = []
actions = [] # 记录操作日志

print(f"开始回测 - 策略: 黄金交叉 (MA5 vs MA10) + 50% 仓位管理")
print(f"初始资金: ${initial_capital:.2f}")
print(f"回测天数: {len(sequential_data)}")
print("-" * 50)

# ==========================================
# 3. 开始循环回测
# ==========================================
for i in range(1, len(sequential_data)):
    # 获取当天和前一天的数据
    today = sequential_data.iloc[i]
    yesterday = sequential_data.iloc[i-1]
    
    current_price = today['open']
    prev_price = yesterday['open']
    
    # ---------------------------------------------------
    # 【熔断机制】检测数据异常 (解决 11元 -> 50元 的问题)
    # ---------------------------------------------------
    price_change_pct = abs(current_price - prev_price) / prev_price
    if price_change_pct > 0.3: # 如果一天内价格跳变超过 30%
        print(f"⚠️ Day {i}: 监测到数据异常跳变 ({prev_price:.2f} -> {current_price:.2f})，判定为更换股票。")
        # 强制清仓，重置持仓，保护利润
        if shares > 0:
            # 假设按前一天收盘价强制结算，而不是今天的开盘价
            cash += shares * yesterday['close']
            print(f"   -> 强制平仓避险。")
            shares = 0
        actions.append('SKIP')
        portfolio_values.append(cash) # 资产不变
        continue # 跳过这一天，不进行交易

    # ---------------------------------------------------
    # 【策略逻辑】黄金交叉 (Golden Cross)
    # ---------------------------------------------------
    ma5 = today['ma_5']
    ma10 = today['ma_10']
    
    # 信号判断
    # 1. 买入信号: MA5 > MA10 (金叉) 且 手里没股票
    buy_signal = (ma5 > ma10)
    
    # 2. 卖出信号: MA5 < MA10 (死叉) 且 手里有股票
    sell_signal = (ma5 < ma10)
    
    # 执行交易
    if buy_signal and shares == 0:
        # 【资金管理】不再 All-in，只用 50% 的现金买入
        position_size = cash * 0.5 
        shares_to_buy = position_size / current_price
        
        # 扣钱，加股票
        cash -= (shares_to_buy * current_price)
        shares += shares_to_buy
        entry_price = current_price
        
        actions.append('BUY')
        print(f"Day {i}: ▲ 买入 {shares_to_buy:.2f} 股 @ ${current_price:.2f} (MA5: {ma5:.2f} > MA10: {ma10:.2f})")
        
    elif sell_signal and shares > 0:
        # 卖出所有持仓
        revenue = shares * current_price
        profit = revenue - (shares * entry_price)
        
        # 加钱，清股票
        cash += revenue
        
        actions.append('SELL')
        color = "💰" if profit > 0 else "🔻"
        print(f"Day {i}: ▼ 卖出 {shares:.2f} 股 @ ${current_price:.2f}. 盈亏: {color} ${profit:.2f}")
        
        shares = 0 # 清零
        
    else:
        actions.append('HOLD')

    # 计算每日总资产
    current_val = cash + (shares * current_price)
    portfolio_values.append(current_val)

# ==========================================
# 4. 结算与绘图
# ==========================================
# 如果最后一天还持有股票，按收盘价折算
final_value = cash + (shares * sequential_data.iloc[-1]['close'])
total_return = (final_value - initial_capital) / initial_capital * 100

print("-" * 50)
print(f"最终资产: ${final_value:.2f}")
print(f"总收益率: {total_return:.2f}%")
print(f"交易次数: {actions.count('BUY')} 买入 / {actions.count('SELL')} 卖出")

# 简单画图看看资产曲线
plt.figure(figsize=(12, 6))
plt.plot(portfolio_values, label='Portfolio Value')
plt.title('Corrected Strategy: Golden Cross (MA5/MA10) with Data Guard')
plt.ylabel('Value ($)')
plt.xlabel('Days')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
